```python
# Load the Pandas libraries with alias 'pd' 
import pandas as pd
from datetime import datetime
import matplotlib.pyplot as plt


data = pd.read_csv("data_adl2mel.csv", index_col=False, parse_dates=["datetime"])
df = pd.DataFrame(data)
#df.set_index('datetime', inplace = True)
#df.index = df.index.tz_localize('Australia/Melbourne').tz_convert('UTC')
# print(df.iloc[:,0].head())
```

```python
#Plot temperature
plt.figure(figsize=(10,5))
plt.plot(df.iloc[:,0], label='Temperatrue')
plt.xlabel('Time, sec')
plt.ylabel('Temperature, ($^\circ$C)')
# plt.xlim(1000,6500)
# plt.ylim(720,1100)
plt.show()
```

```python
#Plot pressure
plt.figure(figsize=(10,5))
plt.plot(df.iloc[:,1], label='Pressure')
plt.xlabel('Time, sec')
plt.ylabel('Pressure, mbar')
# plt.xlim(1000,6500)
# plt.ylim(720,1100)
plt.show()
```

```python
#Plot Humidity
plt.figure(figsize=(10,5))
plt.plot(df.iloc[:,2], label='Humidity')
plt.xlabel('Time, sec')
plt.ylabel('Humidity, %')
# plt.xlim(1000,6500)
# plt.ylim(720,1100)
plt.show()
```

```python
#Plot magnometer X, Y,Z
plt.figure(figsize=(10,5))
plt.plot(df.iloc[:,3], label='Mag X')
plt.plot(df.iloc[:,4], label='Mag Y')
plt.plot(df.iloc[:,5], label='Mag Z')
plt.xlabel('Time, sec')
plt.ylabel('Magnetic field, microT')
plt.legend(loc='upper right')
# plt.xlim(1000,6500)
# plt.ylim(720,1100)
plt.show()
```

```python
#Plot Acceleration X, Y,Z
plt.figure(figsize=(10,5))
plt.plot(df.iloc[:,6], label='Acc X')
plt.plot(df.iloc[:,7], label='Acc Y')
plt.plot(df.iloc[:,8], label='Acc Z')
plt.xlabel('Time, sec')
plt.ylabel('Acceleration, G')
plt.legend(loc='upper right')
# plt.xlim(1000,6500)
# plt.ylim(720,1100)
plt.show()
```

```python
#Plot Orientation X, Y,Z
plt.figure(figsize=(10,5))
plt.plot(df.iloc[:,9], label='Pitch')
plt.plot(df.iloc[:,10], label='Roll')
plt.plot(df.iloc[:,11], label='Yaw')
plt.xlabel('Time, sec')
plt.ylabel('Angle, ($^\circ$)')
plt.legend(loc='upper right')
# plt.xlim(1000,6500)
# plt.ylim(720,1100)
plt.show()
```
